```python
## Python packages - you may have to pip install sqlalchemy, sqlalchemy_utils, and psycopg2.
import pandas as pd
import numpy as np
import datetime 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error as MSE, r2_score as R2S
```

Import data and find how quick employees quit

```python
df = pd.DataFrame.from_csv("employee_retention_data.csv")
df = df.sort_values(['company_id','join_date'])

df['join_date']=pd.to_datetime(df['join_date'])
df['quit_date']=pd.to_datetime(df['quit_date'])

df['daysEmp']=df['quit_date']-df['join_date']
df['daysEmp']=df['daysEmp'].astype(datetime.timedelta).map(lambda x: np.nan if pd.isnull(x) else x.days)

print df.head(1)
print df.describe()
print df.dept.unique()

```

Get new table: day, employee_headcount, company_id
<br>date runs from 2011/01/23 to 2015/12/13

```python
join=df.groupby(['company_id','join_date']).size().reset_index(name='joinCount')
quit=df.groupby(['company_id','quit_date']).size().reset_index(name='quitCount')

df_1=pd.merge(join.rename(columns={'join_date':'day'}), quit.rename(columns={'quit_date':'day'}), how='outer', left_on=['day','company_id'], right_on=['day','company_id'])
df_1=df_1.fillna(0)

df_1 = df_1.sort_values(['company_id','day'])
df_1['joincum']=df_1.groupby(['company_id'])['joinCount'].apply(lambda x:x.cumsum())
df_1['quitcum']=df_1.groupby(['company_id'])['quitCount'].apply(lambda x:x.cumsum())
df_1['headCount']=df_1['joincum'].sub(df_1['quitcum'])

df_1=df_1.drop(['joinCount','quitCount','joincum','quitcum'], 1)
```

Check variation between companies (just for fun)

```python
companies=df_1.groupby('company_id')

#Plot
fig, ax = plt.subplots()
ax.margins(0.05)
for name, group in companies:
    ax.plot(group.day, group.headCount, marker='', linestyle='-', ms=12, label=name)
ax.legend(numpoints=1, loc='upper left')

plt.show()
```
